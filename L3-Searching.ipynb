```python
#Binary Search
def binary_search(arr,x):
    low=0
    high=len(arr)-1
    mid=0
    while low<=high:
        mid=(high+low)//2
        #if x is greater, ignore left half
        if arr[mid]<x:
            low =mid+1
        #if x is smaller, ignore right half
        elif arr[mid]>x:
            high=mid-1
        #means x is present at mid
        else:
            return mid,1
    #if we reach here, then the element was not present
    return mid,0
```

```python
#Monk and Circular Distance
N=int(input())
points=[]
for i in range(N):
    xy= list(map(int,input().split())) 
    d=(xy[0]**2+xy[1]**2)**0.5
    points.append(d)
points.sort()

Q=int(input())
for i in range(Q):
    r=int(input())
    c=0
    for d in points:
        if d<=r:
            c+=1
        else:
            break
    print(c)
```
